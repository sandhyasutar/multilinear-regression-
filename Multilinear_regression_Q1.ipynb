# Calculating VIF

```python
rsq_x = smf.ols('x ~ z + w',data=data).fit().rsquared  
vif_x = 1/(1-rsq_x) # 16.33

rsq_z = smf.ols('z ~ x + w',data=data).fit().rsquared  
vif_z = 1/(1-rsq_z) # 564.98

rsq_w = smf.ols('w ~ x + z',data=data).fit().rsquared  
vif_w = 1/(1-rsq_w) #  564.84

# Storing vif values in a data frame
d1 = {'Variables':['x','z','w'],
      'VIF':[vif_x,vif_z,vif_w]}

Vif_frame = pd.DataFrame(d1)  
Vif_frame
```

# Residual plot for homoscedasticity

```python
def get_standardized_values( vals ):
    return (vals - vals.mean())/vals.std()
```

```python
plt.scatter(get_standardized_values(predictions),
            get_standardized_values(residuals))

plt.title('Residual Plot')
plt.xlabel('Standardized Fitted values')
plt.ylabel('Standardized residual values')
plt.show()
```
